```python
import matplotlib.pyplot as plt
%matplotlib inline
import nclcmaps
import numpy as np
x = y = np.arange(-3.0, 3.01, 0.05)
X, Y = np.meshgrid(x, y)
Z1 = plt.mlab.bivariate_normal(X, Y, 1.0, 1.0, 0.0, 0.0)
plt.pcolormesh(X,Y,Z1,cmap=nclcmaps.cmaps('BlGrYeOrReVi200'))
plt.colorbar()
plt.show()
```
